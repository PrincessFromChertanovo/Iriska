```python
import pandas as pd
pd.options.mode.chained_assignment = None 
names = ['sepal length','sepal width','petal length','petal width','class']
dataframe = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", names = names)
from sklearn.cross_validation import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn import svm 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error 
from math import sqrt 

```

```python
sns.FacetGrid(dataframe, hue= "class", size = 5).map(plt.scatter, "sepal length", "sepal width").add_legend()
sns.plt.show()
```

```python
sns.FacetGrid(dataframe, hue= "class", size = 5).map(plt.scatter, "petal length", "petal width").add_legend()
sns.plt.show()
```

```python
import seaborn as sns
corr = dataframe.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
sns.plt.show()
```

```python
dataframe['class'].replace("Iris-setosa",1,inplace= True)
dataframe['class'].replace("Iris-virginica",2,inplace = True)
dataframe['class'].replace("Iris-versicolor",3,inplace=True)
```

```python
#check for possible values in "class"
dataframe['class'].unique()
```

```python
X = dataframe.iloc[:, 0:4]  
Y = dataframe['class']
```

```python
# X_train -       
# y_train -     
# X_test -    
# y_test -    

X_train, X_test, y_train, y_test = train_test_split(
  X, Y, test_size=0.4, random_state=0)
print(" X_train",X_train)
print("x_test",X_test)
print("y_train",y_train)
print("y_test",y_test)
```

```python
print(X_train.shape, y_train.shape )
print(y_train.shape,y_test.shape)
```

```python
#    
clf = GaussianNB()
clf = clf.fit(X_train ,y_train)
clf.score(X_test, y_test) 

```

```python
#   svm 
clf = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)
clf.score(X_test, y_test)
```
